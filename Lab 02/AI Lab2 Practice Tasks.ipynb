```python
# Practice Task # 1
class SmartLights:
  def __init__(self, name, status):
    self.name = name
    self.status = status

  def switchstatus(self):
    if self.status == 'ON':
      self.status = 'OFF'
      print(f"{self.name} light turned OFF")
    else:
      self.status = 'on'
      print(f"{self.name} light turned ON")

  def displayStatus(self):
      print(f"{self.name} light is {self.status}")

s1 = SmartLights("Living Room", "OFF")
s2 = SmartLights("Bedroom", "ON")
s1.switchstatus()
s2.switchstatus()
s1.displayStatus()
s2.displayStatus()
```

```python
# Practice Task # 2
class Staff:
  def __init__(self, name, id, dept):
    self.name = name
    self.id = id
    self.dept = dept
  def display_info(self):
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.dept}")

class Teacher(Staff):
  def __init__(self, name, id, dept, courses, salary):
    super().__init__(name, id, dept)
    self.courses = courses
    self.salary = salary

  def teach(self):
      print(f"{self.name} is teaching")

  def display_info(self):
    super().display_info()
    print(f"Courses: {self.courses}")
    print(f"Salary: {self.salary}")

class AdminStaff(Staff):
  def __init__(self, name, id, dept, role, working_hours):
    super().__init__(name, id, dept)
    self.role = role
    self.working_hours = working_hours

  def perform_task(self):
      print(f"{self.name} is performing tasks")

  def display_info(self):
    super().display_info()
    print(f"Role: {self.role}")
    print(f"Working Hours: {self.working_hours}")

class ResearchAssistants(Staff):
  def __init__(self, name, id, dept, research_topic, stipend):
    super().__init__(name, id, dept)
    self.research_topic = research_topic
    self.stipend = stipend

  def workonResearch(self):
      print(f"{self.name} is working on {self.research_topic}")

  def display_info(self):
    super().display_info()
    print(f"Research Topic: {self.research_topic}")
    print(f"Stipend: {self.stipend}")

t = Teacher("Saad", 9645, "CS", "DBS", 50000)
a = AdminStaff("Abdullah", 4789, "Admin", "Finance Manager", 7)
r = ResearchAssistants("Hammad", 5621, "SE", "Waterfall", 35000)
t.display_info()
t.teach()
print("\n")
a.display_info()
print("\n")
r.display_info()
r.workonResearch()
```

```python
# Practice Task # 3
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount

  def withdraw(self, amount):
    if self.__balance >= amount:
      self.__balance -= amount
    else:
      print("Insufficient balance")

  def get_balance(self):
    return self.__balance

b1 = BankAccount(4500)
b2 = BankAccount(6200)
b1.deposit(400)
print(f"b1: {b1.get_balance()}")
b2.withdraw(1000)
print(f"b2: {b2.get_balance()}")
b2.withdraw(6000)
```
