```python
# Task 1 – Hill Climbing
# Use the Hill Climbing algorithm to find the maximum value of the function
# f(x) = -x2 + 6x
# Range: 0 ≤ x ≤ 6
# Instructions:
# • Start from a random value of x.
# • Check the neighbor values (x+1 and x−1).
# • Move to the neighbor with the higher value of f(x).
# • Stop when no better neighbor exists.
# Output should show:
# • Initial value of x
# • Value of f(x) at each step
# • Final optimal value.

import random

def f(x):
    return -x**2 + 6*x

def getneighbors(x, xmin=0, xmax=6):
    neighbors = []
    if x-1 >= xmin:
        neighbors.append(x-1)
    if x+1 <= xmax:
        neighbors.append(x+1)
    return neighbors

def hillClimbing(xmin=0, xmax=6):
    currX = random.randint(xmin,xmax)
    currVal = f(currX)

    print(f"Initial x = {currX}, f({currX}) = {currVal}")

    step = 1
    while True:
        neighbors = getneighbors(currX, xmin, xmax)
        nextX = None
        nextVal = currVal

        for neighbor in neighbors:
            neighbourVal = f(neighbor)
            if neighbourVal > nextVal:
                nextX = neighbor
                nextVal = neighbourVal
                break

        if nextX is None:
            print(f"No better neighbor found.")
            break

        currX = nextX
        currVal = nextVal
        print(f"Step {step}: Move to x={currX}, f({currX})={currVal}")
        step = step + 1

    return currX, currVal

print("f(x) = -x^2 + 6x    [0,6]\n")
optimalX, optimalVal = hillClimbing()
print(f"\nFinal Optimal x = {optimalX}")
print(f"Final Optimal f(x) = {optimalVal}")
```

```python
# Task 2 – Beam Search
# Use Beam Search to reach the number 20 starting from 1.
# Allowed operations:
# • x → x + 2
# • x → x + 3
# • x → x × 2
# Heuristic function:
# h(n) = |20 − current number|
# Beam width: k = 2
# Instructions:

# • Start from number 1.
# • Generate the next possible numbers using the allowed operations.
# • At each level keep only the best 2 states with the lowest heuristic value.
# • Continue until the goal number 20 is reached.
# Output should show:
# • Numbers explored at each level
# • Final path to reach 20.

import heapq

def heuristic(n):
    return abs(20-n)

def getsuccessors(x):
    return [x+2, x+3, x*2]

def beamsearch(start, goal, beam_width=2):
    beam = [(heuristic(start), [start])]

    level=0
    while beam:
        print(f"Level {level}: {[path[-1] for _, path in beam]}")
        level = level + 1
        candidates = []

        for h,path in beam:
            current = path[-1]
            if current == goal:
                return path
            for successor in getsuccessors(current):
                new_path = path + [successor]
                candidates.append((heuristic(successor), new_path))

        beam = heapq.nsmallest(beam_width, candidates, key=lambda x: x[0])
    return None

path = beamsearch(start=1, goal=20, beam_width=2)
if path:
    print(f"\nPath found: {' → '.join(map(str, path))}")
else:
    print("\nNo path found.")
```

```python
# Task 3 – Genetic Algorithm
# Use a Genetic Algorithm to maximize the function
# f(x) = x2 + 2x
# Range: 0 ≤ x ≤ 31
# Instructions:
# • Use binary representation for chromosomes.
# • Generate a random initial population.
# • Calculate fitness using the function f(x).
# • Apply selection, crossover, and mutation.
# • Run the algorithm for 15 generations.
# Output should show:
# • Best chromosome
# • Best value of x
# • Best fitness value.

import random
chromosomeLength = 5
populationSize = 6
mutationRate = 0.01
generations = 15

def decode(chromosome):
    return int(''.join(map(str, chromosome)), 2)

def fitness(chromosome):
    x = decode(chromosome)
    return x**2 + 2*x

def createIndividual():
    return [random.randint(0,1) for _ in range(chromosomeLength)]

def selectParent(pop, fitScore):
    total = sum(fitScore)
    pick = random.uniform(0, total)
    current = 0
    for individual, score in zip(pop, fitScore):
        current = current + score
        if current >= pick:
            return individual

def crossover(p1, p2):
    point = random.randint(1, chromosomeLength-1)
    c1 = p1[:point] + p2[point:]
    c2 = p2[:point] + p1[point:]
    return c1, c2

def mutate(chromosome):
    return [bit ^ 1 if random.random() < mutationRate else bit for bit in chromosome]

pop = [createIndividual() for _ in range(populationSize)]
overallBest = None
overallBestFitness = -1

for g in range(1, generations+1):
    fitScore = [fitness(ind) for ind in pop]

    bestIdx = fitScore.index(max(fitScore))
    if fitScore[bestIdx] > overallBestFitness:
        overallBestFitness = fitScore[bestIdx]
        overallBest = pop[bestIdx][:]

    newpop = []
    while len(newpop) < populationSize:
        p1 = selectParent(pop, fitScore)
        p2 = selectParent(pop, fitScore)
        c1, c2 = crossover(p1, p2)
        newpop.append(mutate(c1))
        newpop.append(mutate(c2))

    pop = newpop[:populationSize]

print(f"Best Chromosome: {''.join(map(str, overallBest))}")
print(f"Best x: {decode(overallBest)}")
print(f"Best Fitness: {overallBestFitness}")
```
