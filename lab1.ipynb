```
#q1
stname = input("Enter student name: ")
stmarks = int(input("Enter marks: "))
print("Student Name: ", stname)
print("Marks: ", stmarks)
if stmarks >= 85:
    print("Grade: A")
elif stmarks >= 70:
    print("Grade: B")
elif stmarks >= 50:
    print("Grade: C")
else:
    print("Fail")
```

```
#q2
n = int(input("Enter a number: "))
print("Even numbers: ")
count = 0
for i in range(1,n+1):
    if i%2 == 0:
        print(i)
        count = count + 1
print("Total even numbers: ", count)
```

```
#q3
stdrecords = {}
for i in range(3):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    stdrecords[name] = marks
print("\nStudent Records:")
for name in stdrecords:
    print(name + ":" + str(stdrecords[name]))
```

```
#q4
while True:
    print("\n 1. Add two numbers:\n 2. Subtract two numbers:\n 3. Exit\n")
    choice = int(input("Enter your choice: "))
    if choice == 3:
        print("Program Ended")
        break
    n1 = int(input("Enter first number: "))
    n2 = int(input("Enter second number: "))
    if choice == 1:
        print("Result: ", n1+n2)
    elif choice == 2:
        print("Result: ", n1-n2)
```

```
#q5
def calculate_average(marks):
    mSum=0
    count=0
    for i in range(len(marks)):
        mSum = mSum + marks[i]
        count = count + 1
    avg = mSum/count
    return avg

marks = [78,56,89,91,76]
print("Marks: ", marks)
print("Average Marks: ", calculate_average(marks))
```
