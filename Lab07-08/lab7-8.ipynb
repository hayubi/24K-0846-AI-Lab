```python
# Task 1: Minimax Algorithm Implementation
# 1. Implement the Minimax algorithm for the tree above.
# 2. Compute and print:
# o Minimax values for all nodes
# o Order in which nodes are visited
# 3. Add depth-limiting (depth = 2) and show updated Minimax values.

import math

class Node:
    def __init__(self, value=None):
        self.value = value
        self.children = []
        self.minmax_value = None

class Environment:
    def __init__(self, tree):
        self.tree = tree
        self.visited_order = []

    def compute_minimax(self, node, depth, maximizing_player=True):
        if not node.children:
            self.visited_order.append(node.value)
            return node.value

        if depth == 0:
            self.visited_order.append(node.value)
            return node.value

        if maximizing_player:
            value = -math.inf
            for child in node.children:
                child_value = self.compute_minimax(child, depth - 1, False)
                value = max(value, child_value)

            node.minmax_value = value
            self.visited_order.append(node.value)
            return value

        else:
            value = math.inf
            for child in node.children:
                child_value = self.compute_minimax(child, depth - 1, True)
                value = min(value, child_value)

            node.minmax_value = value
            self.visited_order.append(node.value)
            return value
        
root = Node(0)
n1 = Node(0)
n2 = Node(0)
root.children = [n1, n2]
n3 = Node(0)
n4 = Node(0)
n5 = Node(0)
n6 = Node(0)
n1.children = [n3, n4]
n2.children = [n5, n6]
n3.children = [Node(4), Node(7)]
n4.children = [Node(2), Node(5)]
n5.children = [Node(1), Node(8)]
n6.children = [Node(3), Node(6)]

env = Environment(root)
print("Traversal Order:")
env.compute_minimax(root, depth=3, maximizing_player=True)
print(env.visited_order)
print("\nMinimax Values:")
print("Root:", root.minmax_value)
print("N1:", n1.minmax_value)
print("N2:", n2.minmax_value)
print("N3:", n3.minmax_value)
print("N4:", n4.minmax_value)
print("N5:", n5.minmax_value)
print("N6:", n6.minmax_value)

env2 = Environment(root)
print("\nDepth Limited (Depth = 2):")
env2.compute_minimax(root, depth=2, maximizing_player=True)
print("Traversal Order:", env2.visited_order)
print("\nUpdated Values (Depth 2):")
print("Root:", root.minmax_value)
print("N1:", n1.minmax_value)
print("N2:", n2.minmax_value)
```

```python
# Task 2: Alpha-Beta Pruning
# 1. Implement Minimax with Alpha-Beta pruning on the same tree.
# 2. Track alpha and beta values.
# 3. Print:
# o Minimax values for all nodes
# o Nodes that are pruned
# 4. Compare the number of nodes visited with standard Minimax and explain why
# pruning reduces computation.

import math

class Node:
    def __init__(self, name, value=None):
        self.name = name
        self.value = value
        self.children = []
        self.minmax_value = None


class MinimaxEnv:
    def __init__(self):
        self.visited = []

    def minimax(self, node, maximizing):
        self.visited.append(node.name)

        if not node.children:
            node.minmax_value = node.value
            return node.value

        if maximizing:
            value = -math.inf
            for child in node.children:
                value = max(value, self.minimax(child, False))
            node.minmax_value = value
            return value
        else:
            value = math.inf
            for child in node.children:
                value = min(value, self.minimax(child, True))
            node.minmax_value = value
            return value


class AlphaBetaEnv:
    def __init__(self):
        self.visited = []
        self.pruned = []

    def alpha_beta(self, node, alpha, beta, maximizing):
        self.visited.append(node.name)

        if not node.children:
            node.minmax_value = node.value
            return node.value

        if maximizing:
            value = -math.inf
            for child in node.children:
                child_val = self.alpha_beta(child, alpha, beta, False)
                value = max(value, child_val)
                alpha = max(alpha, value)

                print(f"Node {node.name} (MAX): alpha={alpha}, beta={beta}")

                if beta <= alpha:
                    self.pruned.append(child.name)
                    print(f"Pruned at node: {child.name}")
                    break

            node.minmax_value = value
            return value

        else:
            value = math.inf
            for child in node.children:
                child_val = self.alpha_beta(child, alpha, beta, True)
                value = min(value, child_val)
                beta = min(beta, value)

                print(f"Node {node.name} (MIN): alpha={alpha}, beta={beta}")

                if beta <= alpha:
                    self.pruned.append(child.name)
                    print(f"Pruned at node: {child.name}")
                    break

            node.minmax_value = value
            return value

root = Node("Root")
n1 = Node("N1")
n2 = Node("N2")
root.children = [n1, n2]
n3 = Node("N3")
n4 = Node("N4")
n5 = Node("N5")
n6 = Node("N6")
n1.children = [n3, n4]
n2.children = [n5, n6]
n3.children = [Node("L1",7), Node("L2",4)]
n4.children = [Node("L3",5), Node("L4",2)]
n5.children = [Node("L5",8), Node("L6",1)]
n6.children = [Node("L7",6), Node("L8",3)]

min_env = MinimaxEnv()
min_env.minimax(root, True)

print("\nMINIMAX VALUES:")
print("Root =", root.minmax_value)
print("N1 =", n1.minmax_value)
print("N2 =", n2.minmax_value)
print("\nMINIMAX NODES VISITED:", len(min_env.visited))

ab_env = AlphaBetaEnv()
ab_env.alpha_beta(root, -math.inf, math.inf, True)

print("\nALPHA-BETA VALUES:")
print("Root =", root.minmax_value)
print("N1 =", n1.minmax_value)
print("N2 =", n2.minmax_value)
print("\nALPHA-BETA NODES VISITED:", len(ab_env.visited))
print("\nPRUNED NODES:")
print(ab_env.pruned)

print("\nCOMPARISON:")
print("Minimax visited nodes =", len(min_env.visited))
print("Alpha-Beta visited nodes =", len(ab_env.visited))
print("\nEXPLANATION:")
print("Minimax explores all nodes in the tree.")
print("Alpha-Beta pruning skips branches when alpha >= beta.")
print("This reduces the number of nodes visited.")
print("Therefore, Alpha-Beta improves efficiency without changing the final result.")
```

```python
# Task 3
# 1. Modify the tree by changing leaf values or adding an extra branch.
# 2. Run Minimax and Alpha-Beta again.
# 3. Print:
# o Updated Minimax values
# o Nodes pruned
# o Optimal path for Max from root to leaf
# 4. Comment on any changes in the root value or pruning behavior.

import math

class Node:
    def __init__(self, name, value=None):
        self.name = name
        self.value = value
        self.children = []
        self.minmax_value = None
        self.best_child = None

class Environment:
    def __init__(self, tree):
        self.tree = tree
        self.computed_nodes = []
        self.pruned_nodes = []

    def alpha_beta(self, node, alpha, beta, maximizing):
        self.computed_nodes.append(node.name)

        if not node.children:
            node.minmax_value = node.value
            return node.value

        if maximizing:
            value = -math.inf
            for child in node.children:
                child_val = self.alpha_beta(child, alpha, beta, False)
                if child_val > value:
                    value = child_val
                    node.best_child = child
                alpha = max(alpha, value)
                if beta <= alpha:
                    self.pruned_nodes.append(child.name)
                    break
            node.minmax_value = value
            return value

        else:
            value = math.inf
            for child in node.children:
                child_val = self.alpha_beta(child, alpha, beta, True)
                if child_val < value:
                    value = child_val
                    node.best_child = child
                beta = min(beta, value)
                if beta <= alpha:
                    self.pruned_nodes.append(child.name)
                    break
            node.minmax_value = value
            return value

def get_path(node):
    path = []
    while node:
        path.append(node.name)
        node = node.best_child
    return path

root = Node("Root")
n1 = Node("N1")
n2 = Node("N2")
root.children = [n1, n2]
n3 = Node("N3")
n4 = Node("N4")
n5 = Node("N5")
n6 = Node("N6")
n1.children = [n3, n4]
n2.children = [n5, n6]
n3.children = [Node("L1",7), Node("L2",4)]
n4.children = [Node("L3",5), Node("L4",2)]
n5.children = [Node("L5",8), Node("L6",1)]

extra = Node("N7")
n6.children = [Node("L7",6), Node("L8",3), extra]
extra.children = [Node("L9",9), Node("L10",0)]
env = Environment(root)
result = env.alpha_beta(root, -math.inf, math.inf, True)

print("UPDATED MINIMAX VALUES:")
print("Root =", root.minmax_value)
print("N1 =", n1.minmax_value)
print("N2 =", n2.minmax_value)
print("N3 =", n3.minmax_value)
print("N4 =", n4.minmax_value)
print("N5 =", n5.minmax_value)
print("N6 =", n6.minmax_value)

print("\nNODES PRUNED:")
print(env.pruned_nodes)
print("\nOPTIMAL PATH FOR MAX:")
print(get_path(root))
print("\nTOTAL NODES VISITED:")
print(len(env.computed_nodes))
print("After modifying the tree, the root value may change depending on new leaf values.")
print("Alpha-Beta pruning behavior also changes because pruning depends on node values and order.")
print("Some branches were pruned which reduces computation compared to full Minimax.")
```

```python
# Task 4 Basic CSP Modelling
# Create a CSP model with three variables:
# • Variables: A, B, C
# • Domain: {0,1,2,3}
# Apply the following constraints:
# 1. A ≠ B
# 2. B ≠ C
# 3. A + B ≤ 4
# Requirements:
# • Define variables using model.new_int_var()
# • Add constraints using model.add()
# • Solve the model
# • Print the values of A, B, and C

from ortools.sat.python import cp_model

model = cp_model.CpModel()
A = model.new_int_var(0, 3, 'A')
B = model.new_int_var(0, 3, 'B')
C = model.new_int_var(0, 3, 'C')
model.add(A != B)
model.add(B != C)
model.add(A + B <= 4)

solver = cp_model.CpSolver()
status = solver.solve(model)

if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    print("Solution Found:")
    print("A =", solver.value(A))
    print("B =", solver.value(B))
    print("C =", solver.value(C))
else:
    print("No solution found")
```

```python
# Task 5 Print All Possible Solutions
# Modify the program from Task 1 to print all possible valid solutions.
# Hint:
# Use the solver parameter:
# solver.parameters.enumerate_all_solutions = True
# Also implement a solution callback class to print each solution.
# Output should show all valid assignments of A, B, and C.

from ortools.sat.python import cp_model

class SolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self, A, B, C):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.A = A
        self.B = B
        self.C = C
        self.count = 0

    def on_solution_callback(self):
        self.count += 1
        print(f"Solution {self.count}: A = {self.Value(self.A)}, B = {self.Value(self.B)}, C = {self.Value(self.C)}")

model = cp_model.CpModel()
A = model.NewIntVar(1, 3, 'A')
B = model.NewIntVar(1, 3, 'B')
C = model.NewIntVar(1, 3, 'C')
model.Add(A != B)
model.Add(B != C)
model.Add(A != C)

solver = cp_model.CpSolver()
solver.parameters.enumerate_all_solutions = True
solution_printer = SolutionPrinter(A, B, C)
solver.Solve(model, solution_printer)
print("Total Solutions:", solution_printer.count)
```

```python
# Task 6 Simple Optimization Problem
# Create a CSP model with:
# Variables: x, y, z
# Domain: 0–20
# Constraints:
# x + 2y + z ≤ 20
# 3x + y ≤ 18
# Objective Function:
# maximize 4x + 2y + z
# Requirements:
# • Define variables
# • Add constraints
# • Define the objective function
# • Solve the model
# • Print:
# o Optimal value
# o Values of x, y, and z

from ortools.sat.python import cp_model

model = cp_model.CpModel()
x = model.new_int_var(0, 20, 'x')
y = model.new_int_var(0, 20, 'y')
z = model.new_int_var(0, 20, 'z')
model.add(x + 2*y + z <= 20)
model.add(3*x + y <= 18)
model.maximize(4*x + 2*y + z)

solver = cp_model.CpSolver()
status = solver.solve(model)

if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print("Optimal Value =", solver.objective_value)
    print("x =", solver.value(x))
    print("y =", solver.value(y))
    print("z =", solver.value(z))
else:
    print("No solution found")
```

```python
# Task 7 N-Queens Variation
# Write a Python program using OR-Tools to solve the 4-Queens problem.
# Requirements:
# 1. No two queens share the same row
# 2. No two queens share the same diagonal
# 3. Print at least one valid solution
# Example output format:
# _ Q _ _
# _ _ _ Q
# Q _ _ _
# _ _ Q _

from ortools.sat.python import cp_model

model = cp_model.CpModel()
n = 4
queens = [model.new_int_var(0, n - 1, f'q{i}') for i in range(n)]
model.add_all_different(queens)

for i in range(n):
    for j in range(i + 1, n):
        model.add(queens[i] - queens[j] != i - j)
        model.add(queens[i] - queens[j] != j - i)

solver = cp_model.CpSolver()
status = solver.solve(model)

if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    board = []
    for i in range(n):
        row = ["_"] * n
        row[solver.value(queens[i])] = "Q"
        board.append(row)
    for r in board:
        print(" ".join(r))
else:
    print("No solution found")
```
